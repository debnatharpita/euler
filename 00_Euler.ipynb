```python
# 2
def fibonacci_sequence():
    x, y = 0, 1
    while True:
        x, y = y, x + y
        yield y

f = fibonacci_sequence()
total = 0
while True:
    n = next(f)
    if not n % 2:
        total += n
        
    if n > 4_000_000:
        break

print(total)
```

```python
# 3
import math

def prime_sieve(limit):
    sieve = [True] * (limit + 1)
    sieve[0] = sieve[1] = False
    for i in range(limit + 1):
        if sieve[i]:
            yield i
            
            if i <= math.floor(math.sqrt(limit)):
                for j in range(i * i, limit + 1, i):
                    sieve[j] = False
                
for n in prime_sieve(math.floor(math.sqrt(600_851_475_143))):
    if not 600_851_475_143 % n:
        print(n)
```

```python
# 4
def is_palindrome(n):
    return str(n) == ''.join(reversed(str(n)))

for x in range(999, 899, -1):
    for y in range(x - 1, 899, -1):
        if is_palindrome(x * y):
            print(x, y, x * y)
```

```python
# 5
import math

def prime_factorize(n):
    i = 2
    while i <= n:
        if not n % i:
            yield i
            n //= i
        else:
            i += 1

def pack(l):
    d = {}
    for i in l:
        d[i] = d.get(i, 0) + 1
        
    return d

def unpack(d):
    l = []
    for k, v in d.items():
        l += [k] * v
        
    return l

def max_dict(ds):
    m = {}
    for d in ds:
        for k, v in d.items():
            m[k] = max(m.get(k, 0), v)
    
    return m

factors = [pack(prime_factorize(x)) for x in range(1, 21)]
math.prod(unpack(max_dict(factors)))
```

```python
# 6
def sum_of_squares(l):
    return sum([n * n for n in l])

def square_of_sum(l):
    return sum(l) ** 2

print(square_of_sum(range(1, 101)) - sum_of_squares(range(1, 101)))
```

```python
# 7
def prime_seq():
    yield 2
    
    ps = [2]
    n = 3
    while True:
        prime = True
        for i in ps:
            if not n % i:
                prime = False
                break
        
        if prime:
            yield n
            ps.append(n)
            
        n += 2
            
ps = prime_seq()
for _ in range(10_000):
    next(ps)

print(next(ps))
```

```python
# 8
import math

def digit(ch):
    return ord(ch) - ord('0')

with open('8.txt') as f:
    digits = f.read().replace('\n', '')

m = 1
for i in range(len(digits) - 12):
    m = max(m, math.prod([digit(ch) for ch in digits[i:i + 13]]))
    
print(m)
```

```python
# 11
import math

with open('11.txt') as f:
    numbers = []
    for line in f:
        numbers.append([int(x) for x in line.split(' ')])

m = 0
for row in numbers:
    for i in range(len(row) - 3):
        m = max(m, math.prod(row[i:i + 4]))
        
for i in range(len(numbers) - 3):
    for j in range(len(numbers[0])):
        m = max(m, math.prod([numbers[i][j], numbers[i + 1][j], numbers[i + 2][j], numbers[i + 3][j]]))

for i in range(len(numbers) - 3):
    for j in range(len(numbers[0]) - 3):
        m = max(m, math.prod([numbers[i][j], numbers[i + 1][j + 1], numbers[i + 2][j + 2], numbers[i + 3][j + 3]]))

for i in range(len(numbers) - 3):
    for j in range(3, len(numbers[0])):
        m = max(m, math.prod([numbers[i][j], numbers[i + 1][j - 1], numbers[i + 2][j - 2], numbers[i + 3][j - 3]]))

print(m)
```

```python
# 12
import math

def count_factors(n):
    prime_factors = pack(prime_factorize(n))
    return math.prod([x + 1 for x in prime_factors.values()])

def triangle_numbers():
    i = 1
    while True:
        yield i * (i + 1) // 2
        i += 1

ts = triangle_numbers()
while True:
    n = next(ts)
    if count_factors(n) > 500:
        print(n)
        break
```

```python
# 14
limit = 1_000_001
cache = [0] * limit
cache[1] = 1

def step(n):
    if n % 2:
        return 3 * n + 1
    else:
        return n // 2

def collatz(n):
    if n < limit:
        if not cache[n]:
            cache[n] = 1 + collatz(step(n))

        return cache[n]
    else:
        return 1 + collatz(step(n))

x, y = 0, 0
for n in range(2, limit):
    z = collatz(n)
    if z > x:
        x, y = z, n
    
print(x, y)
```

```python
# 15
def factorial(n):
    f = n
    for i in range(2, n):
        f *= i
        
    return f

print(factorial(40) // (factorial(20) ** 2))
```

```python
# 16
def digital_sum(n):
    if n < 0:
        n = -n
    
    s = 0
    while n:
        s += n % 10
        n //= 10
        
    return s

print(digital_sum(2 ** 1000))
```
